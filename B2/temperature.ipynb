# Import cac thu vien

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

# Them du lieu + xu ly du lieu tho

```python
df = pd.read_csv("temperatures.csv")
df.tail
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
# Chia input/output
f_input = df["Farenheit"].to_numpy()
c_input = df["Farenheit"].to_numpy()
type(f_input), type(c_input)
```

# Huan luyen mo hinh

```python
# Tao random
keras.utils.set_random_seed(42)
```

```python
# Khai bao mo hinh
model = keras.Sequential([
    keras.layers.Dense(units=10, input_shape=[1]),
    keras.layers.Dense(units=1) # output layer
])

model.compile(loss="mean_squared_error", optimizer="adam")
model.summary()
```

```python
# Huan luyen mo hinh
model.fit(f_input, c_input, epochs=100)
```

# Du doan

```python
f_user_input = np.array([90])
# thuc te
c_output_real = round((f_user_input[0] - 32)/1.8, 2)
# du doan
c_output_pred = model.predict(f_user_input)
# ----------------------------------------
print(f"Thuc te: {c_output_real} do C")
print(f"Du doan: {c_output_pred[0][0]:.2f} do C")
```
